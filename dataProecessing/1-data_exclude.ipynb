```python
import pandas as pd
```

```python
# 读取CSV文件
df = pd.read_csv("/data/mr423/project/data/OLINK_data.csv")
```

```python
df.head()
```

```python
df=df.set_index('Unnamed: 0')
df.index.names = ['Id']
```

```python
df.head()
```

```python
baseline_els=[ a  for a in df if (a[-2:] == '_0')]  # eng with 0
df_select = df.loc[:,baseline_els]
df_select.columns=[a[:-2] for a in df_select.columns]
df_select.columns=[a[2:] for a in df_select.columns]

print(df_select.shape)
df_select.head()
```

```python
header = df_select.columns 
header_list = header.tolist() 
unique_processed_header_set = set(header_list)

len(header_list)
```

```python
import matplotlib.pyplot as plt

# 绘制直方图
plt.hist(df_select['EIF4EBP1'], bins=50, range=(-3, 5), edgecolor='k')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

### 个体因缺失 >10% 的蛋白质测量值而被排除

```python
nan_count_per_row = df_select.isna().sum(axis=1)

row_to_drop = nan_count_per_row[nan_count_per_row > len(unique_processed_header_set)*0.1].index
df_drop_item = df_select.drop(row_to_drop)
print(df_drop_item.shape)

df_drop_item.head()
```

```python
nan_count_per_row[row_to_drop]
```

```python
sorted_nan_count_per_row = nan_count_per_row[row_to_drop].sort_values(ascending=False)
sorted_nan_count_per_row
```

```python
# 使用 value_counts() 获取每个值的计数
value_counts = nan_count_per_row[row_to_drop].value_counts()

# 找到出现最多的数量
most_frequent_count = value_counts.idxmax()
most_frequent_count
```
