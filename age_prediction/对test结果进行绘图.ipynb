```python
import scanpy as sc
import numpy as np
import torch
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/data/mr423/project/code/save/biobank-Aug19-11-32/output.csv")
```

```python
df
```

```python
bins = np.arange(30, 100, 10)

plt.figure(figsize=(6, 5)) 
counts, bins, patches = plt.hist(df['Target'], bins=bins, edgecolor='black') 
plt.xlabel('The histogram of Target')
plt.ylabel('Frequency')

for count, bin in zip(counts, bins):
    plt.text(bin + 5, count, str(int(count)), ha='center', va='bottom')


plt.show()
```
