```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import scanpy as sc
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
adata = sc.read("../data/3-OLINK_data_sub_train.h5ad")
adata_test = sc.read("../data/3-OLINK_data_sub_test.h5ad")
```

```python
adata.obs
```

```python
data = pd.read_csv("../data/sub_train_binned_data.csv")
data_test = pd.read_csv("../data/sub_test_binned_data.csv")
```

```python
data=data.set_index('Id')
data_test=data_test.set_index('Id')
```

```python
data
```

```python
data_test
```

找出索引的交集和差异

```python
y_train = adata.obs['age']
y_test = adata_test.obs['age']
```

```python
# # 分割数据为训练集和测试集
# X_train, X_val, y_train, y_val = train_test_split(data, y_train_scaled, test_size=0.3, random_state=0)
```

```python
# 初始化并训练线性回归模型
model = LinearRegression()
model.fit(data, y_train)
```

```python
# 进行预测
y_train_pred = model.predict(data)

# 评估模型性能
train_mse = mean_squared_error(y_train, y_train_pred)
train_mae = mean_absolute_error(y_train, y_train_pred)
train_r2 = r2_score(y_train, y_train_pred)

# 打印评估结果
print("train:")
print(f"均方误差 (MSE): {train_mse:.4f}")
print(f"平均绝对误差 (MAE): {train_mae:.4f}")
print(f"决定系数 (R²): {train_r2:.4f}")
```
